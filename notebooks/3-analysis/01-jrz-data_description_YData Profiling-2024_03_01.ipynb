## ðŸ“š Import  libraries

```python
# base libraries for data science
from pathlib import Path

import pandas as pd
from ydata_profiling import ProfileReport
```

## ðŸ’¾ Load data

```python
DATA_DIR = Path.cwd().resolve().parents[1] / "data"
titanic_df = pd.read_parquet(
    DATA_DIR / "02_intermediate/titanic_type_fixed.parquet", engine="pyarrow"
)
```

## ðŸ“Š Data description

```python
# ordinal data has to be converted again

titanic_df["pclass"] = pd.Categorical(titanic_df["pclass"], categories=[3, 2, 1], ordered=True)

# column Unnamed: 0 is not needed
titanic_df = titanic_df.drop(columns=["Unnamed: 0"])
```

```python
titanic_df.info()
```

```python
type_schema = {
    "embarked": "categorical",
    "pclass": "categorical",
    "sex": "categorical",
}
```

```python
profile = ProfileReport(
    titanic_df,
    title="Titanic Pandas Profiling Report",
    explorative=True,
    correlations={"auto": {"calculate": False}},
    type_schema=type_schema,
)
```

```python
profile.to_widgets()
```

### HTML report embedded in the notebook

```python
profile.to_notebook_iframe()
```

###  ðŸ’¾ Save report to file

```python
profile.to_file(DATA_DIR / "08_reporting/titanic_pandas_profiling_report.html")
```
